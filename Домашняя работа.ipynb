```python
import numpy as np
```

```python
for col in titanic_df.columns:
    Q3 = np.quantile(titanic_df[col], 0.75) 
    Q1 = np.quantile(titanic_df[col], 0.25)
    IQR = Q3 - Q1 
    lower_range = Q1 - 1.5 * IQR 
    upper_range = Q3 + 1.5 * IQR 
    titanic_df = titanic_df[(titanic_df[col] >= lower_range) & (titanic_df[col] <= upper_range)] 

titanic_df
```
